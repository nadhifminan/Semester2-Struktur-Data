```python
def stack():
    s=[]
    return (s)
def push(s,data):
    s.append(data)
def pop(s):
    data=s.pop()
    return(data)
def peek(s):
    return(s[len(s)-1])
def isEmpty(s):
    return (s==[])
def size(s):
    return(len(s))
```

```python
def evaluatePost(postStr):
    operandStack=stack()
    operator='+-/*'
    postStr=postStr.split()
    temp=1
    for i in postStr:
        if i not in operator:
            push(operandStack,i)
        else:
            oprnd2=pop(operandStack)
            oprnd1=pop(operandStack)
            if i=='+':
                result=float(oprnd1)+float(oprnd2)
            elif i=='-':
                result=float(oprnd1)-float(oprnd2)
            elif i=='*':
                result=float(oprnd1)*float(oprnd2)
            elif i == "/":
                result=float(oprnd1)/float(oprnd2)
            print(f'[{temp}]{oprnd1}{i}{oprnd2} = {result}')
            temp+=1
            push(operandStack,result)
    return(pop(operandStack))


print(evaluatePost('10 2 3 + * 4 -'))
```

```python
def createDeque():
    d=[]
    return (d)
def addFront(d,data):
    d.insert(0,data)
    return(d)
def addRear(d,data):
    d.append(data)
    return(d)
def removeRear(d):
    data=d.pop()
    return(data)
def removeFront(d):
    data=d.pop(0)
    return(data)
def isEmpty(d):
    return (d==[])
def size(d):
    return (len(d))
```

```python
def cekPalindrom(kata):
    palindrom=createDeque()
    for ch in kata:
        addFront(palindrom,ch)
    cek=True
    
    while size(palindrom)>1:
        palFront=removeFront(palindrom)
        palRear=removeRear(palindrom)
        print(f'front and rear --> {palFront} == {palRear} ? :{cek}')
        if palFront==palRear:
            cek=cek and True
        else:
            cek=cek and False
    return cek
print(cekPalindrom('katak'))
```

```python
def createQueue():
    q=[]
    return (q)
def enqueue(q,data):
    q.insert(0,data)
    return(q)
def dequeue(q):
    data=q.pop()
    return(data)
def isEmpty(q):
    return (q==[])
def size(q):
    return (len(q))


def sceduling(limitTime,task):
    taskQueue=createQueue()
    for nama in task.keys():
        enqueue(taskQueue,nama)
    #penjadwalan sampai keys habis
    total=0
    while not(isEmpty(taskQueue)):
        temp=dequeue(taskQueue)
        remainTime=task[temp][0]-limitTime
        
        if remainTime>0:
            enqueue(taskQueue,temp)
            procTime=limitTime
        else:
            procTime=task[temp][0]
            remainTime=0
        total=total+procTime
        task[temp][0]=remainTime
        task[temp][1]=total
        
        print(taskQueue)
        
    return task

newTask=sceduling(3,task)
print(newTask)  
```
