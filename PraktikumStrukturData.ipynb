```python
def validateTagHTML(html):
    st=stack()
    html=html.split()
    html=html[1:-1]
    error=[False,'Error :']
    for tag in html:
        if tag[0]=='<' and tag[1]!='/':
            push(st,tag)
        else:
            if tag[0]=='<' and tag[1]=='/':
                if not isEmpty(st):
                    if peek(st)[1:-1]==tag[2:-1]:
                        pop(st)
                    else:
                        error.append(str(len(error)-1)+'. Open tag '+tag+'--> '+peek(st))
                        pop(st)
                else:
                    error.append(str(len(error)-1)+'. kelebihan Close tag : '+tag)
        print(st)
    if len(error)==2 and size(st)==0:
        return True
    else:
        for i in range(size(st)):
            error.append(str(len(error)-1)+'. Kelebihan Open tag : '+pop(st))
        return error

inp="""
<html>
<body>
<title>
Praktikum Struktur Data/
</title>
<h1>
Stack
</h1>
</body>
</html>
"""

validateTagHTML(inp)
```

```python
#buble sort
def bubbleSort(nums,sortBy="Ascending") :
    
    if sortBy == "Ascending" :
        for i in range (len(nums)-1): 
            for j in range(len(nums)-1-i) :
                if nums[j] > nums[j+1] :
                    nums[j],nums[j+1] = nums[j+1],nums[j]
                    
    elif sortBy == "Descending" :
        for i in range (len(nums)-1): 
            
            for j in range(len(nums)-1-i) : 
                if nums[j] < nums[j+1] :
                    nums[j],nums[j+1] = nums[j+1],nums[j]
                
                
    return nums

print(bubbleSort([9,10,1,7] ,"Ascending"))
```

```python
#infix to prefix
import stack as st

def reverseWrd(wrd) :
    wrl = st.stack()
    wrn = ""
    for w in wrd :
        st.push(wrl,w)

    for i in range(len(wrl)) :
        wr = st.pop(wrl)
        if wr == ")" :
            wrn+="("
        elif wr == "(" :
            wrn+=")"
        else :
            wrn+=wr

    return wrn

def convertPrefix(strOp) :
    strOp      = reverseWrd(strOp)
    print(strOp)
    stOperator = st.stack()
    kurung     = ("(",")")
    operator   = {"+":0,"-":0,"*":1,"/":1}
    strRes     = ""
    
    for sto in strOp :
        if sto not in operator.keys() and sto not in kurung :
            strRes=sto+strRes
        
        elif sto == kurung[0] :
            st.push(stOperator,sto)
            
        elif sto == kurung[1] :
            while st.peek(stOperator) != kurung[0] :
                strRes=st.pop(stOperator)+strRes
            st.pop(stOperator)
            
            
        elif sto in operator.keys() :
            while not(st.isEmpty(stOperator)) and st.peek(stOperator) != kurung[0] and operator[st.peek(stOperator)] >= operator[sto] :   
                strRes=st.pop(stOperator)+strRes
            st.push(stOperator,sto)
        
            
    while not(st.isEmpty(stOperator)) :
        strRes=st.pop(stOperator)+strRes
            
            
    return strRes


print("Prefix : ",convertPrefix("2*7+1"))
print("Prefix : ",convertPrefix("2*(7+1)"))
```

```python
#postfix to infix
import stack as st

def postinf(postfix) :
    stOpr    = st.stack()
    operator = ("+","-","*","/")
    
    for pf in postfix :
        if pf not in operator :
            st.push(stOpr,pf)
        else :
            if not(st.isEmpty(stOpr)) :
                opr1   = st.pop(stOpr)
                opr2   = st.pop(stOpr)
                strOpr = "(" + opr2 + pf + opr1 + ")"
                st.push(stOpr,strOpr)
    
    return st.pop(stOpr)

print("Infix = ",postinf("AB*C/DE+-")) 
```

```python
#task scheduling
import queues as que

dataProses = {}
for i in range(3) :
    nama = input("=> Masukkan nama : ")
    dataProses[nama] = int(input("=> Masukkan Waktu : "))
    
def taskScheduling(dictTask,limit) :
    print(dictTask)
    queTask   = que.createQueue()
    queTime   = que.createQueue()
    totalTime = 0
    for dt in dictTask.keys() :
        que.enqueue(queTask,dt)
        
    for dtm in dictTask.values() :
        que.enqueue(queTime,dtm)
   
    while que.size(queTask) > 0 :
        print(queTask)
        limitTemp = limit
        task      = que.dequeue(queTask)
        time      = que.dequeue(queTime)
        
        while limitTemp > 0 :
            limitTemp-=1
            if time > 0 :
                time-=1
                totalTime+=1
        if time > 0 :    
            que.enqueue(queTime,time)
            que.enqueue(queTask,task)
    
    return totalTime
        
print(taskScheduling(dataProses,3))
```
