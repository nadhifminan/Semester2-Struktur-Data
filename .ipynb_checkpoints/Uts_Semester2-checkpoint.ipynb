```python
import stack as st
import deque as dq
import queues as que
#stack
def stack():
    s=[]
    return (s)
def push(s,data):
    s.append(data)
def pop(s):
    data=s.pop()
    return(data)
def peek(s):
    return(s[len(s)-1])
def isEmpty(s):
    return (s==[])
def size(s):
    return(len(s))

#queue
def createQueue():
    q=[]
    return (q)
def enqueue(q,data):
    q.insert(0,data)
    return(q)
def dequeue(q):
    data=q.pop()
    return(data)
def isEmpty(q):
    return (q==[])
def size(q):
    return (len(q))

#deque
def createDeque():
    d=[]
    return (d)
def addFront(d,data):
    d.insert(0,data)
    return(d)
def addRear(d,data):
    d.append(data)
    return(d)
def removeRear(d):
    data=d.pop()
    return(data)
def removeFront(d):
    data=d.pop(0)
    return(data)
def isEmpty(d):
    return (d==[])
def size(d):
    return (len(d))
```

```python
#Prefix to PostFix
def pre2Post(strPre):
    opr = {'*':2,'/':2,'+':1,'-':1}
    oprndStc = stack()
    result = ''
    for i in strPre[::-1]:
        if i not in opr.keys():
            push(oprndStc,i)
        elif i in opr.keys():
            oprnd1=pop(oprndStc)
            oprnd2=pop(oprndStc)
            new = oprnd1 + oprnd2+i
            push(oprndStc,new)
    while not isEmpty(oprndStc):
        result+=pop(oprndStc)
    return result
pre2Post('*+ABC')
# AB+CD+*
```

```python
# Postfix to Infix
def Post2Infix(strPost):
    opr = {'*':2,'/':2,'+':1,'-':1}
    oprndStc = stack()
    result = ''  
    for i in strPost:
        if i not in opr.keys():
            push(oprndStc,i)
        else:
            oprnd1=pop(oprndStc)
            oprnd2=pop(oprndStc)
            math=oprnd2+i+oprnd1
            push(oprndStc,math)
    while not isEmpty(oprndStc):
        result+=pop(oprndStc)
    return result
   
Post2Infix('ABC*+D-')
```

```python
#Postfix to Prefix
def Post2Pref(strPost):
    opr = {'*':2,'/':2,'+':1,'-':1}
    oprndStc = stack()
    result = ''
    for i in strPost:
        if i not in opr.keys():
            push(oprndStc,i)
        elif i in opr.keys():
            oprnd1=pop(oprndStc)
            oprnd2=pop(oprndStc)
            new = i+oprnd2+oprnd1
            push(oprndStc,new)
    while not isEmpty(oprndStc):
        result+=pop(oprndStc)
    return result
print(Post2Pref('AB+C*'))
```

```python
# prefix to Infix
def Pref2Infix(strPost):
    opr = {'*':2,'/':2,'+':1,'-':1}
    oprndStc = stack()
    result = ''  
    for i in strPost[::-1]:
        if i not in opr.keys():
            push(oprndStc,i)
        else:
            oprnd1=pop(oprndStc)
            oprnd2=pop(oprndStc)
            math=oprnd1+i+oprnd2
            push(oprndStc,math)
    while not isEmpty(oprndStc):
        result+=pop(oprndStc)
    return result
   
Pref2Infix('*+AB-CD ')
```

```python
# Infix to Prefix
def inf2Prefix(strMath):
    
    hasil = ""
    oprStc = stack()
    opr = {'*':2,'/':2,'+':1,'-':1}
    count=0
    revStrMath = strMath[::-1]
    
    for i in revStrMath :
        count+=1
        print('read token-',count," : ",i)
        
        if i == ')':
            push(oprStc,i)
            
        elif i == '(':
            while peek(oprStc) != ')':
                hasil+=pop(oprStc)
            pop(oprStc)
            
        elif i in opr.keys():
            while not(isEmpty(oprStc)) and peek(oprStc) != ')' and opr[i] <= opr[peek(oprStc)]:
                hasil+=pop(oprStc)
            push(oprStc,i)
            
        else:
            hasil+=i
            
        print('stack : ',oprStc,"\n")
        
    while not isEmpty(oprStc):
        hasil+= pop(oprStc)
       
    hasil = hasil[::-1]
    return (strMath+" : "+hasil)
    
print(inf2Prefix('A+B-C+D'))
```

```python
#infix to prefix
import stack as st

def reverseWrd(wrd) :
    wrl = st.stack()
    wrn = ""
    for w in wrd :
        st.push(wrl,w)

    for i in range(len(wrl)) :
        wr = st.pop(wrl)
        if wr == ")" :
            wrn+="("
        elif wr == "(" :
            wrn+=")"
        else :
            wrn+=wr

    return wrn

def convertPrefix(strOp) :
    strOp      = reverseWrd(strOp)
    print(strOp)
    stOperator = st.stack()
    kurung     = ("(",")")
    operator   = {"+":0,"-":0,"*":1,"/":1}
    strRes     = ""
    
    for sto in strOp :
        if sto not in operator.keys() and sto not in kurung :
            strRes=sto+strRes
        
        elif sto == kurung[0] :
            st.push(stOperator,sto)
            
        elif sto == kurung[1] :
            while st.peek(stOperator) != kurung[0] :
                strRes=st.pop(stOperator)+strRes
            st.pop(stOperator)
            
            
        elif sto in operator.keys() :
            while not(st.isEmpty(stOperator)) and st.peek(stOperator) != kurung[0] and operator[st.peek(stOperator)] >= operator[sto] :   
                strRes=st.pop(stOperator)+strRes
            st.push(stOperator,sto)
        
            
    while not(st.isEmpty(stOperator)) :
        strRes=st.pop(stOperator)+strRes
            
            
    return strRes


print("Prefix : ",convertPrefix("2*7+1"))
print("Prefix : ",convertPrefix("2*(7+1)"))
```

```python
def preinf(prefix) :
    prefix = reverseWrd(prefix)
    stOpr    = st.stack()
    operator = ("+","-","*","/")
    
    for pf in prefix :
        if pf not in operator :
            st.push(stOpr,pf)
        else :
            if not(st.isEmpty(stOpr)) :
                opr1   = st.pop(stOpr)
                opr2   = st.pop(stOpr)
                strOpr = "(" + opr1 + pf + opr2 + ")"
                st.push(stOpr,strOpr)
    
    return st.pop(stOpr)
print(preinf('/*+ab-cdf'))
```

```python
def convertPostfix(strOp) :
    stOperator = st.stack()
    kurung     = ("(",")")
    operator   = {"+":0,"-":0,"*":1,"/":1}
    strRes     = ""
    
    for sto in strOp :
        if sto not in operator.keys() and sto not in kurung :
            strRes+=sto
        
        elif sto == kurung[0] :
            st.push(stOperator,sto)
            
        elif sto == kurung[1] :
            while st.peek(stOperator) != kurung[0] :
                strRes+=st.pop(stOperator)
            st.pop(stOperator)
            
            
        elif sto in operator.keys() :
            while not(st.isEmpty(stOperator)) and st.peek(stOperator) != kurung[0] and operator[st.peek(stOperator)] >= operator[sto] :   
                strRes+=st.pop(stOperator)
            st.push(stOperator,sto)
        
            
    while not(st.isEmpty(stOperator)) :
        strRes+=st.pop(stOperator)
            
            
    return strRes

infix   = preinf("/*+AB-CDF")
postfix = convertPostfix(infix)

print("Postfix : ",postfix)
```

```python
#buble sort
def bubbleSort(nums,sortBy="Ascending") :
    
    if sortBy == "Ascending" :
        for i in range (len(nums)-1): 
            for j in range(len(nums)-1-i) :
                if nums[j] > nums[j+1] :
                    nums[j],nums[j+1] = nums[j+1],nums[j]
                    
    elif sortBy == "Descending" :
        for i in range (len(nums)-1): 
            
            for j in range(len(nums)-1-i) : 
                if nums[j] < nums[j+1] :
                    nums[j],nums[j+1] = nums[j+1],nums[j]
                
                
    return nums

print(bubbleSort([9,10,1,7] ,"Descending"))
```

```python
def bidirectionalBubbleSort(listData):
    print('Data yang akan diurutkan : ', listData)
    count = 0
    left = 0
    right = len(listData) - 1
    while left < right:
        count += 1
        print('Iterasi ke-', count, ':')
        for i in range(left, right):
            if listData[i] > listData[i+1]:
                listData[i], listData[i+1] = listData[i+1], listData[i]
        print('Kiri-kanan=', listData)

        right -= 1
        for i in range(right, left, -1):
            if listData[i] < listData[i-1]:
                listData[i], listData[i-1] = listData[i-1], listData[i]
        print('Kanan-kiri=', listData)

        left += 1

    return('Data urut-', listData)
a = [10,6,7,1,10,12,100,1,0,23,45,7,8,9]
print(bidirectionalBubbleSort(a))
```

```python
# membalik kata 
def reverse_Kalimat(word):
    dataStack=st.stack()
    for huruf in word:
        st.push(dataStack,huruf)
    hasil=''
    while not(st.isEmpty(dataStack)):
        hasil+=st.pop(dataStack)
        #print(hasil)
    return word,'dibalik menjadi',hasil
print(reverse_Kalimat('nadhif'))
```

```python
def checkParentheses(strMat):
    data=st.stack()
    for ch in strMat:
        if ch=='(':
            st.push(data,ch)
        elif ch==')':
            if st.isEmpty(data):
                return 'kelebihan kurung tutup'
            else:
                st.pop(data)
    if st.isEmpty(data):
        return True
    else:
        return 'Kelebihan kurung buka'
print(checkParentheses('(5 x (4 + 5) / (3 + 2) x (10 - 8))'))
```

```python
def parenthesesCheck(strMath):
    kurung={')':'(','}':'{',']':'['}
    kurungBuka=kurung.values()
    kurungTutup=kurung.keys()
    temp=stack()
    matched=True
    for ch in strMath:
        print('ini adalah ch',ch)
        if ch in kurungBuka: # if ch in kurung.values()
            push(temp,ch)
        if ch in kurungTutup: #if ch in kurung.keys()
            if isEmpty(temp):
                return 'Kelebihan kurung Tutup'
            else:
                tempKurung=pop(temp)
                if tempKurung ==kurung[ch]:#karena dictonary yg diakses menggunakan kurung tutupnya untuk mendapatkan velua kurung bukanya
                    print(tempKurung,'=',kurung[ch])
                    matched=matched and True
                else:
                    matched=matched and False
    if not(isEmpty(temp)):
        return 'Kelebihan Kurung Buka'
    if matched==True:
        return 'OK'
    else:
        return 'Tidak Cocok kurungnya'
    
parenthesesCheck('5+2}*(3)')
```

```python
# POSTFIX=
def operasiPost(str):
    operand_Stack=stack()
    operator='+-/*'
    for i in str:
        if i not in operator:
            push(operand_Stack,i)
        else:
            oprnd2=pop(operand_Stack)
            oprnd1=pop(operand_Stack)
            if i=='+':
                hasil=int(oprnd1)+int(oprnd2)
            elif i=='-':
                hasil=int(oprnd1)-int(oprnd2)
            elif i=='*':
                hasil=int(oprnd1)*int(oprnd2)
            else:
                hasil=int(oprnd1)/int(oprnd2)
            push(operand_Stack,hasil)
    return(pop(operand_Stack))
print(operasiPost('43+52-*'))
```

```python
#queueu
#task scheduling
import queues as que

dataProses = {}
for i in range(3) :
    nama = input("=> Masukkan nama : ")
    dataProses[nama] = int(input("=> Masukkan Waktu : "))
    
def taskScheduling(dictTask,limit) :
    print(dictTask)
    queTask   = que.createQueue()
    queTime   = que.createQueue()
    totalTime = 0
    for dt in dictTask.keys() :
        que.enqueue(queTask,dt)
        
    for dtm in dictTask.values() :
        que.enqueue(queTime,dtm)
   
    while que.size(queTask) > 0 :
        print(queTask)
        limitTemp = limit
        task      = que.dequeue(queTask)
        time      = que.dequeue(queTime)
        
        while limitTemp > 0 :
            limitTemp-=1
            if time > 0 :
                time-=1
                totalTime+=1
        if time > 0 :    
            que.enqueue(queTime,time)
            que.enqueue(queTask,task)
    
    return totalTime
        
print(taskScheduling(dataProses,3))
```

```python
def ularNaga(nama, hitungan):
    gameQueue = createQueue()
    for namaAnak in nama:
        enqueue(gameQueue,namaAnak)
    print('Peserta Permainan=',gameQueue)
    while size(gameQueue) > 1:
        for i in range(hitungan):
            enqueue(gameQueue,dequeue(gameQueue))
            print('hitungan ke-',i,'=',gameQueue)
        dequeue(gameQueue)
        print('Peserta Permainan=',gameQueue)
    return dequeue(gameQueue)
ularNaga(['andi','rita','sari','anton','rafa','diana','zaki'],2)
```
